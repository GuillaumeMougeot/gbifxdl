```python
def get(request):
    get_template = "https://api.gbif.org/v1/species/match?name={}".format(request)
    species_dict = json.loads(requests.get(get_template).text)
    if "usageKey" in species_dict.keys():
        return species_dict["usageKey"]
    else:
        return "NOT FOUND"
```
